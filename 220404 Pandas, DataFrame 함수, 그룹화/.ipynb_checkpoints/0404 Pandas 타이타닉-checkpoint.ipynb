```python
import pandas as pd
# 타이타닉 데이터
import seaborn as sns
titanic = sns.load_dataset('titanic')
titanic
```

```python
# 칼럼별 결측치 갯수
titanic.isna().sum()
```

```python
# 기본 정보
titanic.info()
```

```python
# 결측치를 제외한 갯수
titanic.count()
```

```python
titanic.head()
```

```python
# who 열의 child 행의 갯수
titanic.loc[titanic['who']=='child'].count()
```

```python
# 상관관계 (-1에 가까울수록 반비례, 1에 가까울수록 정비례)
titanic.corr()['survived']
```

```python
# age 결측치
b = titanic['age'].isna()
age_na_idx = titanic[b].index
titanic.loc[age_na_idx,:]
```

```python
# age의 평균
titanic['age'].mean()
```

```python
# 평균값을 결측치에 대입(원본수정) => 2가지방법
titanic['age'] = titanic['age'].fillna(29.6)
titanic['age'].fillna(29.6, inplace=True)
```

```python
titanic_male = titanic["sex"]=='male'
titanic_female = titanic.drop(titanic[titanic_male].index)
titanic_female
```

```python
titanic2 = titanic.drop(['embark_town','class','adult_male','deck','alive'], axis=1)
titanic2
```

```python
# 항목에 있는 요소로 그룹을 나누는 함수 (groupby)
titanic.groupby("sex")
```

```python
# 항목별 평균
titanic.groupby("pclass")["survived"].mean()
```

```python
# 성별에서 탑승클래스별 생존률 (여성만 추출)
ss = titanic.groupby(['sex','pclass'])["survived"].mean()
ss['female']
```

```python
# 계층적 index
# 여성중 1클래스 탑승객의 생종률
sp = titanic.groupby(['sex','pclass'])["survived"].mean()
sp.index
ss[('female', 1)]
```

```python
# 생존여부와 나이에 대한 평균, 최대값 출력 (agg)
titanic.groupby(['sex','pclass'])[["survived", "age"]].agg(["mean","max"])
```
