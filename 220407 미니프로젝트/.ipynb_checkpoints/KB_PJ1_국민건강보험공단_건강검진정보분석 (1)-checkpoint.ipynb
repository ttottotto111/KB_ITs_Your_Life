```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', 31)

plt.rcParams['font.family'] = 'Malgun Gothic' # 한글 폰트 설정
plt.rcParams['figure.figsize'] = (10, 6) # 그래프 크기 설정
```

```python
df=pd.read_csv("./국민건강보험공단_건강검진정보_20211229.CSV",encoding='euc-kr')
df
```

```python
df[df['허리둘레'].isna()]
```

```python
df=df.reset_index().drop('index',axis=1)
```

# 1. 비만 분석
- BMI, WWI, 허리둘레

```python
df['BMI']=df['체중(5Kg 단위)']/(df['신장(5Cm단위)']/100).pow(2)
df['WWI']=df['허리둘레']/df['체중(5Kg 단위)'].pow(1/2)
```

```python
df['비만']='정상체중'
df.loc[df['BMI']>26,'비만']='과체중'
df.loc[df['BMI']>30,'비만']='고도비만'
```

```python
df['복부비만']=0
df.loc[(df['성별코드']==1)&(df['허리둘레']>=90),'복부비만']=1
df.loc[(df['성별코드']==2)&(df['허리둘레']>=85),'복부비만']=1
```

```python
df_size = df.pivot_table(index="비만", columns="복부비만", aggfunc="size")
df_size
```

```python
sns.heatmap(df_size, cmap=sns.light_palette("magenta", as_cmap=True), annot=True, fmt="d")
plt.title("Heatmap")
plt.show()
```

# 2. 혈압측정

```python
df['혈압']='정상'
df.loc[df['이완기 혈압']>=90,'혈압']='고혈압'
df.loc[df['수축기 혈압']>=140,'혈압']='고혈압'
df.loc[df['이완기 혈압']<=50,'혈압']='저혈압'
df.loc[df['수축기 혈압']<=80,'혈압']='저혈압'
```

```python
df=pd.concat([df,pd.get_dummies(df['혈압'])],axis=1)
df.drop(['혈압','정상'],inplace=True,axis=1)
```

# 3. 빈혈 유무

```python
df['빈혈']=0
```

```python
df.loc[(df['혈색소']<=13)&(df['성별코드']==1),'빈혈']=1
df.loc[(df['혈색소']<=12)&(df['성별코드']==2),'빈혈']=1
```

# 4. 혈중 지질 검사

```python
df.columns
```

```python
df['지질 검사 결과']=0
```

```python
df.loc[df['총 콜레스테롤']>=200,'지질 검사 결과']=1
df.loc[df['LDL 콜레스테롤']>=130,'지질 검사 결과']=1
df.loc[df['HDL 콜레스테롤']<=40,'지질 검사 결과']=1
df.loc[df['트리글리세라이드']>=150,'지질 검사 결과']=1
```

```python
df['지질 검사 결과'].value_counts()
```

# 5. 당뇨 검사

```python
df['당뇨']=0
```

```python
df.loc[df['식전혈당(공복혈당)']>=126,'당뇨']=1
```

```python
df['당뇨'].value_counts()
```

# 6. 간 기능 검사

```python
df['간 기능']=0
```

```python
df.loc[df['(혈청지오티)AST']>=40,'간 기능']=1
df.loc[df['(혈청지오티)ALT']>=40,'간 기능']=1
df.loc[(df['감마 지티피']>=70)&(df['성별코드']==1),'간 기능']=1
df.loc[(df['감마 지티피']<=10)&(df['성별코드']==1),'간 기능']=1
df.loc[(df['감마 지티피']>=42)&(df['성별코드']==2),'간 기능']=1
df.loc[(df['감마 지티피']<=6)&(df['성별코드']==2),'간 기능']=1
```

# 7.신장기능검사

```python
df['신장 기능']=0
```

```python
df.loc[(df['혈청크레아티닌']<=0.7)|(df['혈청크레아티닌']>=1.4),'신장 기능']=1
```
