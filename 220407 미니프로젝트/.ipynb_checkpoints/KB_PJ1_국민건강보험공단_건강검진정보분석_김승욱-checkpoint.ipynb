# 각 코드를 값으로 변경

```python
# 성별
def sex(x):
    if x ==1 or x=="남자": return "남자"
    if x==2 or x=="여자": return "여자"

# 흡연
def smok(x):
    if x==1 or x=="비흡연": return "비흡연"
    if x==2 or x=="금연": return "금연"
    if x==3 or x=="흡연": return "흡연"

# 음주
def drink(x):
    if x==0 or x=="N" or x=="음주 X": return "음주 X"
    if x==1 or x=="Y" or x=="음주 O": return "음주 O"
    
# 나이대
def year(x):
    if x==9 or x==10: return 40
    if x==11 or x==12: return 50
    if x==13 or x==14: return 60
    if x==15 or x==16: return 70
    if x==17 or x==18: return 80

df["연령대"] = None
df["성별코드"] = df["성별코드"].apply(sex)
df["흡연상태"] = df["흡연상태"].apply(smok)
df["음주여부"] = df["음주여부"].apply(drink)
df["연령대"] = df["연령대 코드(5세단위)"].apply(year)
df
```

# 1. 비만 분석
- BMI, WWI, 허리둘레

```python
df['BMI']=df['체중(5Kg 단위)']/(df['신장(5Cm단위)']/100).pow(2)
df['WWI']=df['허리둘레']/df['체중(5Kg 단위)'].pow(1/2)
```

```python
df['비만']='정상체중'
df.loc[df['BMI']>26,'비만']='과체중'
df.loc[df['BMI']>30,'비만']='고도비만'
```

```python
df['복부비만']="정상"
df.loc[(df['성별코드']=="남자")&(df['허리둘레']>=90),'복부비만']="비만"
df.loc[(df['성별코드']=="여자")&(df['허리둘레']>=85),'복부비만']="비만"
```

```python
df_size = df.pivot_table(index="비만", columns="복부비만", aggfunc="size")
df_size
```

```python
sns.heatmap(df_size, cmap=sns.light_palette("magenta", as_cmap=True), annot=True, fmt="d")
plt.title("Heatmap")
plt.show()
```
