```python
import matplotlib.pyplot as plt
import numpy as np

# x축은 인덱스
# y축은 값

x = np.arange(10)
plt.plot(x**2)

# x,y축의 범위 설정
plt.xlim(0,90)
plt.xlim(0,90)
```

```python
#한글 세팅
plt.rcParams['font.family']="Malgun Gothic"

# x, y값을 지정, 모양지정
score = [70,80,90,60,70]
plt.plot(range(5),score,'ro')

score2 = [80,90,70,80,100]
plt.plot(range(5),score2,'b--')

score3 = [60,30,40,10,60]
plt.plot(range(5),score3,'c^')

#타이틀
plt.title("한글 타이틀")

plt.ylim(0,100)
```
