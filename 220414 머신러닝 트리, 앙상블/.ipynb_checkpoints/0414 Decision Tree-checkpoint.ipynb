```python
import pandas as pd
import seaborn as sns
from sklearn import tree
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import DecisionTreeRegressor
from sklearn.tree import export_graphviz
```

### Decision Tree

```python
cancer = load_breast_cancer()

X = cancer['data']
Y = cancer['target']
x_tr, x_te, y_tr, y_te = train_test_split(X,Y,random_state=0, stratify=Y)

treemodel = DecisionTreeClassifier().fit(x_tr, y_tr)
treemodel.score(x_tr, y_tr),treemodel.score(x_te, y_te)
```

```python
#비교
Logimodel = LogisticRegression(max_iter=10000).fit(x_tr, y_tr)
Logimodel.score(x_tr, y_tr),Logimodel.score(x_te, y_te)
```

```python
#비교
KNmodel = KNeighborsClassifier().fit(x_tr, y_tr)
KNmodel.score(x_tr, y_tr),KNmodel.score(x_te, y_te)
```

```python
treemodel.feature_importances_
```

```python
# 특성 중요도 시각화
fi = pd.Series(treemodel.feature_importances_, index = cancer['feature_names'])
fi[fi!=0].sort_values(ascending=False).plot(kind = 'bar', figsize=(10,5))
```

```python
# 트리 그래프로 출력
plt.figure(figsize=(20,20))
tree.plot_tree(treemodel, fontsize=10)
plt.show()
```

```python
export_graphviz(treemodel, out_file="treemodel.dot",
               feature_names=cancer.feature_names,
               class_names= cancer.target_names,
               impurity=True, filled=True)
```

```python
# 과대 적합인 경우 트리의 깊이를 정해 해결
# 일반적으로 max_depth가 낮을수록 좋음
treemodel = DecisionTreeClassifier(max_depth=3).fit(x_tr, y_tr)
treemodel.score(x_tr, y_tr),treemodel.score(x_te, y_te)
```

```python
# 결정 트리 회귀 알고리즘 사용

boston = load_boston()

X = boston['data']
Y = boston['target']
x_tr, x_te, y_tr, y_te = train_test_split(X,Y,random_state=0)

treemodel = DecisionTreeRegressor().fit(x_tr, y_tr)
treemodel.score(x_tr, y_tr),treemodel.score(x_te, y_te)
```

```python
fi = pd.Series(treemodel.feature_importances_, index = boston['feature_names'])
fi[fi!=0].sort_values(ascending=False).plot(kind = 'bar', figsize=(10,5))
```

```python
# tree 시각화
plt.figure(figsize=(20,20))
tree.plot_tree(treemodel)
plt.show()
```
