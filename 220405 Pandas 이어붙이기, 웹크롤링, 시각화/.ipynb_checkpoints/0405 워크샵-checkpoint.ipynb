# 1. 데이터 불러오기

```python
first = pd.read_csv('gas_first_2019.csv', encoding='euc-kr')
second = pd.read_csv('gas_second_2019.csv', encoding='euc-kr')
```

```python
# 상반기 판매가격 데이터 프레임 결측치 조회
first.isna().sum()
```

```python
# 하반기 판매가격 데이터 프레임 결측치 조회
second.isna().sum()
```

# 2. 데이터 결합

```python
# 상반기와 하반기 데이터를 상하로 결합
df = pd.concat([first, second]).reset_index()
df
```

# 3. 데이터 가공 및 분석

```python
# 데이터 프레임 정보 조회
df.info()
```

```python
# 상표 컬럼의 값(value) 분포 조회 
df["상표"].value_counts()
```

```python
# 지역 컬럼의 값(value) 분포 조회 
df["지역"].value_counts()
```

tip.
1. apply를 활용합니다.
2. split을 활용합니다.

예시)
<br>df['지역'].apply(split_place).str[0] ==> 앞단어
<br>df['지역'].apply(split_place).str[0] ==> 뒷단어

```python
# 지역 컬럼의 값을 이용하여 시와 구를 분리
df["시"] = df['지역'].str.split(" ").str[0]
df["구"] = df['지역'].str.split(" ").str[1]
df.head()
```

```python
# 시 컬럼의 서울을 서울특별시로 변환
def f1(x):
    if x =="서울": return "서울특별시"
df["시"] = df["시"].apply(f1)
df.head()
```

```python
# 휘발유 가격이 비싼 5개 판매가격을 출력하세요.
df.sort_values("휘발유", ascending=False).head(5)
```

```python
# 휘발유 가격이 저렴한 5개 판매가격을 출력하세요. (단 휘발유가 0원인 데이터는 제외)
df[df["휘발유"]!=0].sort_values("휘발유").head(5)
```

```python
# 상표별 평균 휘발유 가격을 계산하세요.
df.groupby("상표", as_index=False)[["휘발유"]].agg("mean")
```

```python
# 상표와 셀프여부별 평균 휘발유 가격을 계산하세요.
df.groupby(["상표", "셀프여부"])[["휘발유"]].agg("mean")
```

```python
# 지역별 평균 휘발유 가격을 계산하세요
df.groupby("구")["휘발유"].agg("mean")
```

```python
# pivot_table 함수를 이용하여 구와 상표 컬럼을 이용하여 평균 휘발유 가격을 계산하세요
df.pivot_table(index = "구",  
                columns="상표", 
               values = "휘발유",
                aggfunc='mean')
```
