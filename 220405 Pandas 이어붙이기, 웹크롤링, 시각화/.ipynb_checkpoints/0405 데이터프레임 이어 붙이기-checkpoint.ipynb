```python
import pandas as pd
import seaborn as sns
titanic = sns.load_dataset("titanic")
titanic
```

```python
# apply 함수
def f1(x):
    if x == "man": return "남자"
    if x == "woman": return "여자"
    if x =="child": return "아이"

titanic["who"].apply(f1)
```

```python
# apply함수 lamda와 같이 사용
titanic["who"].apply(lambda x: "남자" if x=="man" else "여자" if x=="woman" else "아이").value_counts()
```

```python
def f2(x):
    print(x.shape)
    
#titanic.apply(f2) #칼럼단위 시리즈 전달
titanic.apply(f2,axis=1) #행단위 시리즈가 전달
```

```python
# apply 보기쉽게 정리
score = pd.DataFrame(
    {
        "이름":["홍", "김"],
        "영어":[70, 90],
        "수학":[50,80]
    }
)
score
```

```python
# 열을 확장하는 방법
def calc_avg(x):
    return x.mean()

score[["영어", "수학"]].apply(calc_avg) # 열 기준
score["평균"] = score[["영어", "수학"]].apply(calc_avg, axis=1) # 행 기준
score
```

```python
score1 = pd.DataFrame(
    {
        "이름":["홍", "김"],
        "중간영어":[70, 90],
        "중간수학":[50,80]
    }
)
score2 = pd.DataFrame(
    {
        "학생이름":["홍", "김"],
        "기말영어":[70, 90],
        "기말수학":[50,80]
    }, 
    index=[3,4]
)

pd.concat([score1, score2]) #밑으로 붙임
#pd.concat([score1, score2], axis=1, join="inner") #옆으로 붙임 (인덱스 기준)
```

```python
# 칼럼이 다르면 오류
pd.merge(score1, score2)
```

```python
# 칼럼이 다를경우 lefton righton
pd.merge(score1, score2, left_on="이름", right_on="학생이름")
```
