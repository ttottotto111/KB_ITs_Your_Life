```python
import pandas as pd

apt_price = pd.read_csv(
    "./아파트(매매)__실거래가_20220401113837.csv",
    skiprows=15,
    encoding="euc-kr",
    thousands=",")
apt_price
```

```python
apt_price["평"] = apt_price["전용면적(㎡)"]/3.03
apt_price.head()
```

```python
# 계약년월을 연도와 월로 나눔
apt_price["계약년"] = apt_price["계약년월"]//100
apt_price["계약월"] = apt_price["계약년월"]%100
apt_price.head()
```

```python
# 칼럼의 항목이 문자열이어야 한다 (str)
# str => 대신 접근함
apt_price["시군구"].str.split()
```

```python
# 한자리 날짜는 01,02~ 로 표현하기위해 where로 0과 결합
# 문자열로 변경후 결합
import numpy as np
apt_price["계약일"] = np.where(apt_price["계약일"]<10,
        "0"+apt_price["계약일"].astype(str),
        apt_price["계약일"].astype(str))
apt_price["계약년월일"] = apt_price["계약년월"].astype(str)+apt_price["계약일"].astype(str)
```

```python
# 년월일 데이터를 datetime으로 날짜 정도 수정
apt_price["계약년월일"] = pd.to_datetime(apt_price["계약년월일"])
apt_price["계약년월일"]
```

```python
# 년월일 데이트타임으로 바뀐정보 확인
apt_price.info()
```

```python
# 요일 출력 (dayofweek)
apt_price["계약년월일"].dt.dayofweek
```

```python
# 값을 경계값을 기준으로 자르는 함수 (cut)
apt_price["평형"]=pd.cut(
    apt_price["평"],
    bins = [0,10,20,30,40,50,60,1000], #구간 경계값
    labels=['10평이하','10평대','20평대','30평대','40평대','50평대','60평이상'] # 구간별 카테고리
) 
apt_price.head()
```

```python
apt_price["평형"].value_counts()
```

```python
# 거래금액
apt_price["가격대"] = pd.cut(
    apt_price["거래금액(만원)"],
    bins=[0,10000,100000,200000,300000,400000,500000,1000000],
    labels=["1억이하","10억대","20억대","30억대","40억대","50억대","50억이상"])
apt_price.head()
```

```python
import time
import pandas as pd
import numpy as np

def get_weather_info(year,month=None):
    '''year, month의 날씨를 크롤링해서 반환.
    '''
    pass
    if month == None:
        start_month = 1
        end_month = 13
    else:
        start_month = month
        end_month = month + 1

    import time

    #날짜별 날씨 저장용
    weathers = [] 

    for month in range(start_month, end_month):
        time.sleep(0.5)
        url=f"https://www.weatheri.co.kr/bygone/pastDB_tmp.php?jijum_id=108&start={year}-{month}-01&end={year}-12-31"
        print(url)

        weather = pd.read_html( url )

        w = weather[3]
        w.columns = w.loc[0,:]
        w = w[["날짜","날씨"]]
        w.drop(0, axis=0,inplace=True)

        def get_date(x):
            return x[0:4] + x[6:8] + x[9:11]

        w.loc[ : , "날짜"] = w["날짜"].apply( get_date )
#         display(w.head())
        weathers.append(w)
    # 추출한 날씨를 하나로 합치기
    w_all = weathers[0]
    for i in weathers[1:]:
        w_all = pd.concat([w_all, i], ignore_index=True)
    return w_all

if __name__ == "__main__":
    #테스트코드
    print(get_weather_info(2021,3))
```

```python
# 강남3구외 그외지역

# 강남 3구만 추출
b_gn3 = apt_price["구"].isin(["강남구", "서초구", "송파구"])
# 강남 3구외 추출 (~ => not)
b_gn22 = ~apt_price["구"].isin(["강남구", "서초구", "송파구"])

b20 = apt_price["평형"]=="20평대"


# 강남 3구이면서 20평대인 데이터 추출
gn3_b20 = apt_price.loc[b_gn3 & b20, :]
# 강남3구외의 20평대의 평과 거래금액 추출
gn3_b22 = apt_price.loc[b_gn22 & b20, :]

# 시각화
plt.plot(gn3_b20["평"],
         gn3_b20["거래금액(만원)"],
         'bo',
         label = "강남3구"
    )
plt.plot(gn3_b22["평"],
         gn3_b22["거래금액(만원)"],
         'r^',
         label = "강남3구 이외"
    )
```

```python
# 강남 3구와 그 외 지역의 거래금액(만원) 의 describe()함수 결과를 합친 데이터 프레임 출력
apt_price[b_gn3][["거래금액(만원)"]].describe()
apt_price[b_gn22][["거래금액(만원)"]].describe()
gn = pd.concat([gn3_b20[["거래금액(만원)"]].describe(),gn3_b22[["거래금액(만원)"]].describe()],axis = 1)

plt.bar(["강남3구 거래평균", "그외지역"],
       gn.loc["mean",:])
```
