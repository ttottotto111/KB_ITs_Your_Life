```python
# urllib => urlretrieve를 사용한 이미지 처리
import urllib.request
import os.path
url="http://www.careshop.co.kr/eimages/editor/Filedata_20160302111675636.jpg"
basename = os.path.basename(url)
basename = url.split("/")[-1]
print("basename", basename)
urllib.request.urlretrieve(url, basename)
print('End')
```

```python
# 파라미터 전달법(Parameter)
# 기존 주소 : https://www.kma.go.kr/wid/queryDFSRSS.jsp?zone=1168064000
import requests
url = "https://www.kma.go.kr/wid/queryDFSRSS.jsp"
params = {'zone': 1168064000}

r = requests.get(url=url, params=params)
print(r.text)
print(r.url)
```

```python
# 파싱 외부 라이브러리 (BeautifulSoup)
# 찾는 방법 1.태그이름으로 찾기 2.find(옵션) 3.select(css선택자)

# find함수 -> 맨위 하나의 태그만 찾는함수
from bs4 import BeautifulSoup

soup = BeautifulSoup(r.text, "html.parser")

ctg = soup.find("title")
print(ctg.text)
```

```python
# find_all 해당태그를 모두 찾는 함수
titles = soup.find_all("title")
display(titles)
print(titles[0].text)
```

```python
# 원하는 데이터 추출하기

#<data>. . .</data> 단위로 추출
data_arr = soup.find_all("data")
print(len(data_arr))

for dt in data_arr:
    h = dt.find("hour").text
    t = dt.find("temp").text
    d = dt.find("day").text
    wf = dt.find("wfkor").text
    print(h,t,d,wf)
```

# 환율 파싱

```python
import requests
from bs4 import BeautifulSoup
url = "https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=0&ie=utf8&query=%ED%99%98%EC%9C%A8"
r = requests.get(url=url)

soup = BeautifulSoup(r.text, "html.parser")

tr_arr = soup.select("table.rate_table_info>tbody>tr")

환율 = dict()

for tr in tr_arr:
    # 문자열은 불변 객체
    c = tr.select("td")[0].text.replace(",","")
    환율[tr.select("th")[0].text] = float(c)
display(환율)
```

```python
# 교보문고 베스트셀러 추출하기 예제
import requests
from bs4 import BeautifulSoup
url = "http://www.kyobobook.co.kr/bestSellerNew/bestseller.laf"
r = requests.get(url=url)

soup = BeautifulSoup(r.text, "html.parser")

book = soup.select("ul.list_type01 div.title strong")

bestbook = []
for b in book:
    bestbook.append(b.text)
display(bestbook)
```
