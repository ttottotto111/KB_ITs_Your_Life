- email: 이메일 소유 여부
- occyp_type: 직업 유형													
- family_size: 가족 규모
- begin_month: 신용카드 발급 월
    - 데이터 수집 당시 (0)부터 역으로 셈, 즉, -1은 데이터 수집일 한 달 전에 신용카드를 발급함을 의미

## Binary variables
- gender
- car
- reality

```python
df.head()
```

```python
# 'gender' 열의 값을 'F'는 0, 'M'은 1로 바꾸고 빈도수를 구하세요.
def gender(x):
    if x== 'F':
        return 0
    else:
        return 1
df['gender']=df['gender'].apply(gender)
df['gender'].value_counts()
```

```python
# 'car' 열의 값을 'N'는 0, 'Y'은 1로 바꾸고 빈도수를 구하세요.
def car(x):
    if x== 'N':
        return 0
    else:
        return 1
df['car']=df['car'].apply(car)
df['car'].value_counts()
```

```python
# 'reality' 열의 값을 'N'는 0, 'Y'은 1로 바꾸고 빈도수를 구하세요.
def reality(x):
    if x== 'N':
        return 0
    else:
        return 1
df['reality']=df['reality'].apply(reality)
df['reality'].value_counts()
```

## 학습, 검증 데이터 분리

```python
df.head()
```

```python
# 'credit' 열을 제외한 나머지 열은 X, 'credit' 열은 y 에 저장하세요.
X = df.drop('credit',axis=1)
y = df['credit']
```

```python
# 학습 데이터와 검증 데이터를 3 : 1 비율로 분리하세요. (y 값 빈도수 비율을 유지)
# random_state = 42
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25,random_state=42)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

<br/>
<br/>
<br/>

## 모델링
- Decision Tree
- Random Forest
- xgboost
- lightgbm

```python
# Decision Tree 모델을 학습한 뒤 Train, Test 데이터의 score를 구하세요.
# 하이퍼 파라미터를 자유롭게 조절하세요.
from sklearn.tree import DecisionTreeClassifier

treemodel = DecisionTreeClassifier(max_depth=15).fit(X_train, y_train)
treemodel.score(X_train, y_train),treemodel.score(X_test, y_test)
```

```python
# Random Forest 모델을 학습한 뒤 Train, Test 데이터의 score를 구하세요.
# 하이퍼 파라미터를 자유롭게 조절하세요.
from sklearn.ensemble import RandomForestClassifier

RFmodel = RandomForestClassifier(max_depth=16).fit(X_train, y_train)
RFmodel.score(X_train, y_train),RFmodel.score(X_test, y_test)
```

```python
# xgboost 모델을 학습한 뒤 Train, Test 데이터의 score를 구하세요.
# 하이퍼 파라미터를 자유롭게 조절하세요.
from xgboost import XGBClassifier
xgbmodel = XGBClassifier(objective='multi:softprob')
xgbmodel.fit(X_train, y_train)

xgbmodel.score(X_train, y_train),xgbmodel.score(X_test, y_test)
```

```python
# lightgbm 모델을 학습한 뒤 Train, Test 데이터의 score를 구하세요.
# 하이퍼 파라미터를 자유롭게 조절하세요.
from lightgbm import LGBMClassifier
lgbmodel = LGBMClassifier()
lgbmodel.fit(X_train, y_train)

lgbmodel.score(X_train, y_train),lgbmodel.score(X_test, y_test)
```

```python
# lightgbm이 X_test 를 예측한 결과를 pred에 저장하세요.
pred = lgbmodel.predict(X_test)
pred
```

```python
# 정답과 비교하여 confusion matrix를 구하세요. (그대로 실행)
from sklearn.metrics import confusion_matrix

pd.DataFrame(confusion_matrix(y_test, pred),
             index = [['Real']*3, [0,1,2]],
             columns = [['Predict']*3, [0,1,2]])
```
