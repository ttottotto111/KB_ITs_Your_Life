# 중고차 판매 정보 분석 실습

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
# vehicles.csv 파일을 읽어 df 변수에 저장하세요.
df = pd.read_csv("./vehicles.csv")
df
```

```python
# Unnamed: 0 컬럼을 제거하세요.
df = df.drop("Unnamed: 0", axis=1)
df
```

```python
# 가격 컬럼의 단위를 만원으로 변경하세요.
df["가격"] = df["가격"]/10000
df
```

```python
# 제조사별 판매 차량 수를 출력하세요.
df["제조사"].value_counts()
```

```python
# 제조사별 판매 차량 수를 파이차트로 시각화하세요.
plt.figure(figsize = (10,10))

x = df["제조사"].value_counts()
labels = plt.pie(x, labels=x.index,autopct="%.2f%%")
plt.show()
```

```python
# 유형별 판매 차량 수를 출력하세요.
df["유형"].value_counts()
```

```python
# 유형별 판매 차량 수를 bar 차트로 시각화하세요.
plt.figure(figsize=(10, 7))
a = df.groupby("유형")[["모델"]].count()
a = a.sort_values(by="모델", ascending=False)
x = "유형"
y = "판매 수"
sns.barplot(data =a, x=a.index,y="모델")

plt.show()
```

```python
# 판매 가격이 가장 높은 데이터 6개를 출력하세요.
df.sort_values(by="가격", ascending=False).head(6)
```

```python
# 판매 가격이 가장 높은 6개 제품을 '제조사:모델' 대비 가격의 형태로 bar 차트 시각화하세요.
plt.figure(figsize=(10, 7))

top6 = df.sort_values(by="가격", ascending=False).head(6)
y = "가격"

sns.barplot(data=top6, x="제조사":"모델", y=y,orient='h')





plt.show()
```
