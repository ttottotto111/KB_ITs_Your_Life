- 조건(condition) 인덱싱

```python
# '구분' 열의 고유값을 조회하세요.
df["구분"].unique()
```

```python
# '구분' 열의 값이 '무인카페' 인 행을 조회하세요.
df[df["구분"]=='무인카페']
```

```python
# '구분' 열의 값에 'DT' 가 포함된 행을 조회하세요.
df[df['구분'].str.contains('DT')]
```

```python
# '구분' 열의 값에 '무인' 이 포함된 행을 조회하세요.
df[df['구분'].str.contains('무인')]
```

<br/>
<br/>

- 2가지 이상의 조건(condition) 인덱싱

```python
# '매출건수' 가 '유효회원수' 의 3배보다 많은 행을 조회하세요.
df[df['매출건수']>df['유효회원수']*3]
```

```python
# 패스트푸드DT를 이용하는 남성의 행을 조회하세요.
cond1 = df[df['구분'].isin(['패스트푸드DT'])]
cond2 = cond1[cond1['성별'].isin(['남성'])]
cond2
```

```python
# 20대 여성의 행을 조회하세요.
cond1 = df[df['연령대'].isin(['20대'])]
cond2 = cond1[cond1['성별'].isin(['여성'])]
cond2
```

```python
# 20, 30대 중 유효회원수가 30000 이상인 행을 조회하세요.
cond1 = df[df['연령대'].isin(['20대','30대'])]
cond2 = cond1[cond1['유효회원수']>30000]
cond2
```

```python
# 유효회원수가 30000 이상이거나 매출건수가 유효회원수의 3배 이상인 행을 조회하세요.
cond1 = df['유효회원수']>30000
cond2 = df['매출건수']>df['유효회원수']*3
df[cond1|cond2]
```
