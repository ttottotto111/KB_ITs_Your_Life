```python
# 업무추진비 크롤링(파일 읽기 함수)
import pandas as pd
url = "https://raw.githubusercontent.com/seoul-opengov/opengov/master/expense_list2021/202101_expense_list.csv"
seoul_exp01 = pd.read_csv(url)
seoul_exp01
```

```python
# 데이터프레임 생성
mem01 = pd.DataFrame([
    [70,80,90],
    [90,80,80]
])
mem01.index = ["홍","김"]
mem01.columns = ["언어","수학","과학"]
mem01
```

```python
# 그래프 (x축 기준)
# 한글지원을 안하기때문에 오류
mem01.T.plot()
```

- df[]
- 열 추출 : df[칼럼] => 인덱싱, df[[칼럼,칼럼]] => 팬시인덱싱(여러가지 값 출력), 
- 행 추출 : df[인덱스:인덱스]

- df.loc[행 인덱스] => 행 추출, df.loc[행 인덱스, 칼럼] => 값 1개 추출, df.loc[:,칼럼] => 칼럼 추출

```python
# unique() 중복제거, shape 요소 개수
b1 = seoul_exp01['dept_nm_lvl_1']=="서울시본청"
b1
se1 = seoul_exp01[b1] # True에 해당하는 행 추출.
se1['dept_nm_lvl_1'].unique()
```

```python
# 원본을 바꾸는함수 inplace=True / 결과가 나오면 복사본이 바뀐것, 안나오면 원본이 바뀐것
mem01.rename(columns={"언어":"영어2"}, inplace=True)
mem01
```

```python
# skiprows => ~몇째 줄까지 스킵, thousands ="," => 천자리마다 로 구분되어 있는 항목 정수로 변환
seoul_apt = pd.read_csv(
    "./아파트(매매)__실거래가_20220401113837.csv", 
    encoding="EUC-KR", 
    skiprows=15, 
    thousands=",")
seoul_apt.info()
```

```python
# df.loc[행, 열] => 인덱서
# 0행 "단지명"에 있는 요소
seoul_apt.loc[0,"단지명"]
```

```python
#0번행 전체 출력
# loc Series로 표현 (1차원으로 출력)
#seoul_apt.loc[0]
#loc[[~],~]`=> 데이터칼럼으로 표현 (2차원으로 출력)
# 슬라이싱 할때 끝번호 포함 (이하)
seoul_apt.loc[2:5,"단지명":"거래금액(만원)"]
```

```python
# 여자인데 생존한 사람의 연령만 출력
titanic.loc[2:5,"class":"deck"]
b = (titanic["sex"]=="female") & (titanic["survived"]==1)
titanic.loc[b,"age"]
```
