```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['font.family'] = 'Malgun Gothic' # 한글 폰트 설정
plt.rcParams['figure.figsize'] = (10, 6) # 그래프 크기 설정
```

```python
rate = pd.read_csv("./rate.csv")
rate.dropna(inplace=True)
rate = rate.reset_index()
rate.drop("index", axis = 1, inplace=True)
rate
```

```python
rrr = pd.read_csv("./rrr2.csv", encoding='cp949')
rrr.drop("Unnamed: 0", axis = 1, inplace = True)
rrr
```

```python
df = pd.read_csv("./attack_rate.csv")
df
```

```python
# 세트별 평균

setdata = rate.groupby("여자부")[["1SET","2SET","3SET","4SET","5SET"]].agg("mean")
setdata.drop(index = "페퍼저축은행", inplace=True)
setdata
```

```python
att_dep = df.groupby("팀명").mean()[["공격 효율", "수비"]]
att_dep.drop(index = "페퍼저축은행", inplace=True)
att_dep = att_dep.reset_index()
att_dep
```

## 시각화

```python
sns.pairplot(data = df.drop(["경기날짜","팀명", "1SET","2SET","3SET","4SET","5SET"], axis=1))
plt.show()
```

```python
sns.histplot(data = df.drop(["경기날짜","팀명", "1SET","2SET","3SET","4SET","5SET"], axis=1))
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.subplot(1,2,1)
sns.barplot(data = att_dep.sort_values("공격 효율", ascending=False), x="팀명", y="공격 효율")
plt.subplot(1,2,2)
sns.barplot(data = att_dep.sort_values("수비", ascending=False), x="팀명", y="수비")
```
