```python
# 데이터프레임 변경 (날짜, 팀 별 출전 선수 명단과 결과)
df_starting5 = df.groupby(['경기날짜', '팀명'])[['이름', '결과']].agg({'이름' : lambda data : ' '.join(data).split(), '결과' : any})
df_starting5 = pd.DataFrame(df_starting5).reset_index()
df_starting5['이름'] = df_starting5['이름'].str[:5]

# 경기번호 추가
total = df_starting5.shape[0] // 2
df_starting5['경기번호'] = np.array([[i,i] for i in range(1, total+1)]).flatten()
df_starting5
```

```python
rate['경기번호'] = np.array([[i,i] for i in range(1, total+1)]).flatten()
rate
```

```python
# 데이터프레임 변경 (경기 별 두 팀의 이름, 출전 선수 명단과 결과)
def concat_names(data) :
    result = []
    for d in data :
        result += d
    return result
df_game = df_starting5.groupby('경기번호').agg({'경기날짜' : lambda dates : dates.values[0],
                                                '팀명' : lambda teams : ' vs '.join(teams),
                                                '이름' : concat_names,
                                                '결과' : np.argmax}).reset_index()
df_game
```

```python
# 선수의 직전 기록 찾는 함수 정의
# 이전 3개의 경기에서 등비수열을 이용해 예측값 저장
def find_previous_record(df, date, name) :
    day1 = pd.Timedelta(days=1)
    date -= day1
    count=0
    tmp=[]
    while date >= df.loc[0, '경기날짜'] :
        row = df[(df['경기날짜'] == date) & (df['이름'] == name)]
        if len(row) : # 이전 기록 존재
            tmp.append(row.iloc[0, :])
            count+=1;
            if count==3 : break
        date -= day1
    
    if count==1:
        c=tmp[0]
        return c
    elif count==2:
        c=tmp[0]
        c[4:13]=0.7*(tmp[0][4:13])+0.3*(tmp[1][4:13])
        c[14]=0.7*(tmp[0][14])+0.3*(tmp[1][14])
        c[-2]=0.7*(tmp[0][-1])+0.3*(tmp[1][-2])
        c[-1]=0.7*(tmp[0][15])+0.3*(tmp[1][-1])
        return c
    elif count==3:
        c=tmp[0]
        c[4:13]=0.7*(tmp[0][4:13])+0.21*(tmp[1][4:13])+0.09*(tmp[2][4:13])
        c[14]=0.7*(tmp[0][14])+0.21*(tmp[1][14])+0.09*(tmp[2][14])
        c[-1]=0.7*(tmp[0][-1])+0.21*(tmp[1][-1])+0.09*(tmp[2][-1])
        c[-2]=0.7*(tmp[0][-2])+0.21*(tmp[1][-2])+0.09*(tmp[2][-2])
        return c
    else:
        return []
```

```python
# 각 선수 별 직전 기록을 새로운 열에 추가
cols = ['공격종합_성공률','오픈_성공률','시간차_성공률','이동_성공률',
        '후위_성공률','속공_성공률','퀵오픈_성공률','서브_성공률','범실_범실', 
        '공격 효율', '수비','0','1','2','3','4']

rows = []
for i in range(len(df_game)) :
    date = df_game.loc[i, '경기날짜']
    names = df_game.loc[i, '이름']
    row = []
    
    for name in names :
        result = find_previous_record(df, date, name)
        if len(result) == 0 :
            row += [np.nan] * len(cols)
        else :
            row += list(result[cols].values.flatten())
    rows.append(row)

    
columns = []
for team in ['A팀', 'B팀'] :
    for player in ['선수1', '선수2', '선수3', '선수4', '선수5'] :
        columns += [f'{team}_{player}_{col}' for col in cols]
        
df_game = pd.concat([df_game, pd.DataFrame(rows, columns = columns)], axis = 1)
df_game.tail()
```
