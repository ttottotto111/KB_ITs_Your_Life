```python
from pycaret.classification import *
from sklearn.datasets import load_iris

iris = load_iris()

iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df['target'] = iris['target']
iris_df
```

```python
iris_df_setup = setup(data = iris_df,target = 'target')
```

```python
compare_models()
```

```python
# 최적의 모델 3개 추출
top3 = compare_models(n_select=3, sort = "Accuracy")
top3
```

```python
# 3개의 모델중 가장 최적의 하이퍼 파라미터 출력
top3_tuned = [tune_model(n) for n in top3]
top3_tuned
```

```python
top3_blend = blend_models(top3)
top3_blend
```

```python
final_model = finalize_model(top3_blend)
final_model
```

```python
# target = 원래값, label = 예측값
pred = predict_model(final_model, data = iris_df.iloc[-20:])
pred
```

```python
# 예측률
from pycaret.utils import check_metric
check_metric(pred.target, 
             pred.Label, 
             metric = "Accuracy")
```

```python
# 시각화
plot_model(
    top3_blend,
    plot="residuals"
)
```
