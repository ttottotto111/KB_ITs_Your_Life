```python
import numpy as np

x1 = np.array([10,9,3,2])
y1 = np.array([90,80,50,30])

x2 = np.array([10,9,3,2,11])
y2 = np.array([90,80,50,30,40])
```

## 이상치 처리
1. 표준 점수로 변환 후 제거
- 평균이 0, 표준편차가 1인 분포로 변환 후 3 이상 -3 이하인 경우 이상치로 판단
2. IQR 이용 제거
- 사분위값을 이용해 75~25% 계산후 1.5를 곱하여 처리

## 정규화
- 값을 0~1 사이로 줄이는 것

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))

# 이상치 X
plt.subplot(1,2,1)
plt.scatter(x1,y1)

# 이상치 포함
plt.subplot(1,2,2)
plt.scatter(x2,y2)
```

```python
# 이상치 제거 함수

def OLS(x,y):
    w = np.sum((x-x.mean())*(y-y.mean()))/np.sum((x-x.mean())**2)
    b = y.mean()-w*x.mean()
    return w, b
```

```python
w1, b1 = OLS(x1,y1)
w1, b1
```

```python
w2, b2 = OLS(x2, y2)
w2, b2
```

```python
plt.figure(figsize=(10,5))

# 이상치 X
plt.subplot(1,2,1)
plt.scatter(x1,y1)
plt.plot(x1, x1*w1+b1, c="red")

# 이상치 포함
plt.subplot(1,2,2)
plt.scatter(x2,y2)
plt.plot(x2, x2*w2+b2, c="red")
```

## Linear Regression (회귀)
- 선형 상관 관계를 모델링하는 회귀 분석
1. 최소 제곱법
2. 경사 하강법
- 미분을 통해 기울기를 얻음,  
기울기에 하강속도를 추가하여 가중치 값에서 빼는것 반복  
=> 답에 가까울수록 기울기가 작아지고, 답에 멀수록 기울기가 커짐  
(기울기가 임계치보다 작으면 답이라고 판단)

```python
# 최소 제곱법 사용
from sklearn.linear_model import LinearRegression

model1 = LinearRegression().fit(x1.reshape(-1,1), y1)
model1.coef_, model1.intercept_
```

```python
model2 = LinearRegression().fit(x2.reshape(-1,1), y2)
model2.coef_, model2.intercept_
```

```python
# 보스턴 주택 가격 예측
import pandas as pd

url = "https://raw.githubusercontent.com/vincentarelbundock/Rdatasets/master/csv/MASS/Boston.csv"

boston = pd.read_csv(url)
boston.drop("Unnamed: 0", axis = 1, inplace=True)

x = boston.iloc[:,:-1]
y = boston.iloc[:,-1]

from sklearn.model_selection import train_test_split

#stratify => 분류를 예측하는것 / 수치 예측 X
x_tr, x_te, y_tr, y_te = train_test_split(x,y, random_state=0, test_size=0.2)

# 모델 학슴
from sklearn.linear_model import LinearRegression

model = LinearRegression().fit(x_tr, y_tr)
model.score(x_tr,y_tr),model.score(x_te, y_te)
```

```python
# 칼럼별 기울기
model.coef_, model.intercept_
```

```python
# 그래프
import matplotlib.pyplot as plt

plt.scatter(boston.iloc[:,0], 
            boston.iloc[:,-1])

plt.plot(boston.iloc[:,0],
         boston.iloc[:,0]*model.coef_[0]+model.intercept_, 
         c="red")
```

```python
# 그래프
import matplotlib.pyplot as plt

for i in range(x.shape[1]):
    pass
    plt.figure(i)
    plt.scatter(boston.iloc[:,i], 
                boston.iloc[:,-1])

    plt.plot(boston.iloc[:,i],
             boston.iloc[:,i]*model.coef_[i]+model.intercept_, 
             c="red")
```

```python
import numpy as np

np.random.seed(0)
X = np.arange(50)
Y = (2*X)+10*np.random.randn(50)
plt.plot(X,Y,'b.')
```

## 값에서 멀수록 기울기가 커지고 가까울수록 기울기가 작아짐
### => 확률적 경사 하강법

```python
# 가설(경사 하강법)
w=0
np.mean((w*X-Y)**2)
```

```python
# 가설(경사 하강법)
w=1
np.mean((w*X-Y)**2)
```

```python
w_range = np.arange(0.1,4.1,0.1)
costs=[]

for w in w_range:
    h = w*X
    cost = 1/50 * np.sum((h-Y)**2)
    costs.append(cost)
plt.plot(w_range, costs, 'r.')
```

```python
plt.plot(w_range, costs, 'r.')
for w, cost in zip(w_range, costs):
    h=w*X
    gradient = 2/50 * np.sum((h-Y)*X)
    plt.plot(w_range, gradient*(w_range-w)+cost)
    plt.axis([0,4,0,3000])
plt.show()
```

```python
# 경사 하강법을 적용
from sklearn.linear_model import SGDRegressor

model = SGDRegressor(max_iter = 10000).fit(x_tr,y_tr)
model.score(x_te, y_te)
```

```python
# 경사 하강법을 적용 
# fit => 1차원 데이터는 사용 불가
# verbose : 로그를 찍어주는 함수
from sklearn.linear_model import SGDRegressor

model = SGDRegressor(max_iter = 10000, verbose=1).fit(X.reshape(-1,1),Y)
model.score(X.reshape(-1,1),Y)
```

## Polynomia features
- 선형이 아닌경우 1차식으로 표현이 불가능해,  
2차식 이상의 다항식으로 변형하여 회귀 분석을 진행 할 때 사용  
ex) P = PolynomialFeatures(degree=2)

```python
# polynomia features 다항 회귀
# (a+b)**2 = a**2 + b**2 + 2ab => 2ab 상호교차항

# x0를 2차항으로 변환

from sklearn.preprocessing import PolynomialFeatures

# 제곱 degree=2
P = PolynomialFeatures(degree=2)
x = np.arange(5).reshape(-1,1)

#feature => 칼럼
pd.DataFrame(P.fit_transform(x), columns=P.get_feature_names())
```

```python
#x0,x1특성을 2차항 변환

# 제곱 degree=2
P = PolynomialFeatures(degree=2)
x = np.arange(10).reshape(-1,2)

#feature => 칼럼
pd.DataFrame(P.fit_transform(x), columns=P.get_feature_names())
```

```python
#include bias 상수항을 빼고 싶을때

# 제곱 degree=2
P = PolynomialFeatures(degree=2, include_bias=False)
x = np.arange(10).reshape(-1,2)

#feature => 칼럼
pd.DataFrame(P.fit_transform(x), columns=P.get_feature_names())
```

```python
# 3차항 => degree =3
P = PolynomialFeatures(degree=2, include_bias=False)
x_poly = P.fit_transform(boston.iloc[:,:-1])
y = boston.iloc[:,-1]
x_tr, x_te, y_tr, y_te = train_test_split(x_poly,y,random_state=0)

boston.loc[:,'crim':'lstat'].shape, x_poly.shape
```

```python
model = LinearRegression().fit(x_tr, y_tr)
model.score(x_tr, y_tr), model.score(x_te, y_te)
```

## 규제화
- 다항 회귀시 차수가 높을수록 과대적합이 발생  
=> 가중치의 크기제어 필요 (가중치 = 기울기)
1. Ridge Regression (L2규제 가중치)
- 가중치의 제곱의 합을 주는것 (대부분 L2사용)  
값을 크게 줄수록 기울기가 일정해짐
2. Lasso Regression (L1 규제 가중치)
- 가중치의 절대값의 합을 주는것  
중요도가 크면 커지고 작으면 작아짐.  
많이주면 원하는값이 안나올 수 있음  
영향력이 작은 값을 0으로 만들어 없앨때 사용

```python
# 규제화 (L2 규제화)
from sklearn.linear_model import Ridge
model = Ridge(alpha=10).fit(x_tr, y_tr)
model.score(x_tr, y_tr), model.score(x_te, y_te)
```
