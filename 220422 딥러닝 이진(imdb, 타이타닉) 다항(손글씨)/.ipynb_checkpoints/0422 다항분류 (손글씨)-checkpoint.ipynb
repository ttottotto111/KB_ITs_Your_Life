## 다항분류 실습

```python
import tensorflow as tf
import numpy as np

mnist = tf.keras.datasets.mnist.load_data()
(x_tr, y_tr), (x_te, y_te) = mnist
print(x_tr.shape, y_tr.shape)
print(x_te.shape, y_te.shape)

print(y_tr[:5])
import matplotlib.pyplot as plt

for i in range(5):
    plt.subplot(1,5,i+1)
    plt.imshow(x_tr[i], cmap="gray")
```

### 2차원을 1차원으로 변경

```python
(x_tr, x_te) = (np.float32(x_tr)/255, np.float32(x_te)/255)

n_input = 28*28
(x_tr, x_te) = (x_tr.reshape(-1,n_input),x_te.reshape(-1,n_input))
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(300, activation="relu", input_shape=(28*28,)),
    tf.keras.layers.Dense(100, activation="relu"),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(10, activation="softmax")
])
model.summary()
model.compile(loss = "sparse_categorical_crossentropy",
             optimizer = 'adam',
             metrics=['accuracy'])

model.fit(x_tr, y_tr,
         batch_size = 100,
         epochs=5,
         validation_split=0.2)

test_scores = model.evaluate(x_te, y_te, verbose=2)

print('Test loss:', test_scores[0])
print('Test accuracy:', test_scores[1])
```

## 직접 쓴 숫자 인식

```python
import numpy as np
import cv2
import img2data

img_path = './0458.png'

image = cv2.imread(img_path)
plt.imshow(image)
plt.show()

numbers =img2data.img2digits(image, (28,28), border=4)

for i, n in enumerate(numbers):
    print(n.shape)
    
    Z = model.predict(n.astype(np.float32)/255.0)
    
    pred = np.argmax(Z, axis=1)
    print(Z, pred)
    plt.subplot(1, len(numbers), i+1)
    plt.title(str(pred))
    plt.imshow(n.reshape(28,28), cmap="gray")
```
